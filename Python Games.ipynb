```python
import random
```

```python

# 1: Riddle Logic Puzzle Game

riddles = {
    "I speak without a mouth and hear without ears. I have no body, but I come alive with wind. What am I?": "echo",
    "I have cities, but no houses; I have mountains, but no trees; I have water, but no fish. What am I?": "map",
    "What is seen in the middle of March and April that can't be seen at the beginning or end of either month?": "r",
    "What belongs to you, but other people use it more than you do?": "your name",
    "I am always hungry and will die if not fed, but whatever I touch will soon turn red. What am I?": "fire",
    "What word is spelled incorrectly in every dictionary?": "incorrectly",
    "What has many teeth, but can’t bite?": "comb",
    "What can fill a room but takes up no space?": "light",
    "If you drop me I’m sure to crack, but give me a smile and I’ll always smile back. What am I?": "mirror",
    "What gets wetter the more it dries?": "towel"
}

def riddle_game():
    score = 0
    total = 0
    asked = set()
    print("Welcome to the Riddle Logic Puzzle Game!")
    print("Type 'quit' anytime to exit.\n")
    while len(asked) < len(riddles):
        question, answer = random.choice(list(riddles.items()))
        if question in asked:
            continue
        asked.add(question)
        print("Riddle:", question)
        user_answer = input("Your answer (or type 'quit' to exit): ").lower().strip()
        if user_answer == 'quit':
            break
        total += 1
        if user_answer == answer or user_answer in answer.split():
            print("Correct!\n")
            score += 1
        else:
            print("Incorrect. The correct answer is:", answer, "\n")
    if total > 0:
        print(f"Your score: {score}/{total} = {score/total*100:.2f}%")
    else:
        print("No riddles were attempted.")


# 2: Arithmetic Logic Puzzle Game

def arithmetic_puzzle():
    score = 0
    total = 0
    print("Welcome to the Arithmetic Logic Puzzle Game!")
    print("Type 'quit' at any time to exit.\n")
    while True:
        num1 = random.randint(1, 20)
        num2 = random.randint(1, 20)
        operator = random.choice(['+', '-', '*', '/'])


        if operator == '/':
            num2 = random.randint(1, 10)
            correct_answer = round(num1 / num2, 2)
            question = f"{num1} / {num2}"
        elif operator == '+':
            correct_answer = num1 + num2
            question = f"{num1} + {num2}"
        elif operator == '-':
            correct_answer = num1 - num2
            question = f"{num1} - {num2}"
        else:
            correct_answer = num1 * num2
            question = f"{num1} * {num2}"

        print(f"Solve: {question}")
        user_input = input("Your answer (or type 'quit' to exit): ")
        if user_input.lower().strip() == 'quit':
            break

        try:
            user_answer = float(user_input)
            total += 1
            if round(user_answer, 2) == round(correct_answer, 2):
                print("Correct!\n")
                score += 1
            else:
                print(f"Incorrect. The correct answer is: {correct_answer}\n")
        except ValueError:
            print("Invalid input. Please enter a number.\n")
    if total > 0:
        print(f"Your score: {score}/{total} = {score/total*100:.2f}%")
    else:
        print("No puzzles were attempted.")


# 3: General Knowledge Quiz Game

categories = {
    "Science": [
        {"question": "Which gas do plants absorb during photosynthesis?", "answer": "carbon dioxide"},
        {"question": "What is the full name of the WWW?", "answer": "world wide web"},
        {"question": "How many bones are there in the adult human body?", "answer": "206"},
    ],
    "Mathematics": [
        {"question": "What is the only even prime number?", "answer": "2"},
        {"question": "What is the sum of the first 5 natural numbers?", "answer": "15"},
        {"question": "What is the name of the mathematical symbol π?", "answer": "pi"},
    ],
    "Geography": [
        {"question": "What is the tallest mountain on Earth?", "answer": "mount everest"},
        {"question": "What is the capital of France?", "answer": "paris"},
        {"question": "Which planet is known as the 'Red Planet'?", "answer": "mars"},
    ],
    "Others": [
        {"question": "Which country won the most recent FIFA World Cup (as of 2022)?", "answer": "argentina"},
        {"question": "Who invented the light bulb?", "answer": "thomas alva edison"},
        {"question": "In what year did India declare itself a Democratic and Republic state with the adoption of the Constitution?", "answer": "1950"},
    ]
}

def ask_category():
    print("\nCategories:")
    for idx, cat in enumerate(categories, 1):
        print(f"{idx}. {cat}")
    while True:
        try:
            choice = int(input("Select a category by number: "))
            if 1 <= choice <= len(categories):
                return list(categories.keys())[choice - 1]
            else:
                print("Please enter a number corresponding to the categories above.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def ask_question(q):
    print("\nQuestion:", q["question"])
    answer = input("Your answer: ").strip().lower()
    return answer

def check_answer(user_answer, correct_answer):
    if user_answer == correct_answer or user_answer in correct_answer.split():
        print("Correct!\n")
        return True
    else:
        print(f"Incorrect. The correct answer is: {correct_answer.capitalize()}\n")
        return False

def display_score(total, score):
    percent = (score/total)*100 if total > 0 else 0
    print(f"Quiz Complete! Your score: {score}/{total} = {percent:.2f}%\n")

def exec_questions(questions):
    score = 0
    random.shuffle(questions)
    for q in questions:
        user_ans = ask_question(q)
        if check_answer(user_ans, q["answer"]):
            score += 1
    display_score(len(questions), score)

def play_gk_quiz():
    print("Welcome to the General Knowledge Quiz Game!")
    while True:
        cat = ask_category()
        print(f"\nYou have selected: {cat}")
        exec_questions(categories[cat])
        again = input("Do you want to play again? (yes/no): ").strip().lower()
        if again != 'yes':
            break
    print("Thank you for playing!")


# 4: Rock-Paper-Scissors (Player vs Computer)

def get_computer_choice():
    return random.choice(['rock', 'paper', 'scissors'])

def get_user_choice():
    choice = input("Enter your choice (rock, paper, scissors) or 'quit' to exit: ").lower().strip()
    while choice not in ['rock', 'paper', 'scissors', 'quit']:
        print("Invalid choice. Try again.")
        choice = input("Enter your choice (rock, paper, scissors) or 'quit' to exit: ").lower().strip()
    return choice

def decide_winner(computer, user):
    if computer == user:
        return "tie"
    wins = {
        'rock': 'scissors',
        'scissors': 'paper',
        'paper': 'rock'
    }
    if wins[computer] == user:
        return "computer"
    else:
        return "user"

def play_rock_paper_scissors_computer():
    print("Welcome to Rock-Paper-Scissors ")
    user_score = 0
    computer_score = 0
    ties = 0

    while True:
        user_choice = get_user_choice()
        if user_choice == 'quit':
            print("\nThanks for playing! Final scores:")
            print(f"You: {user_score} |  Computer: {computer_score} | Ties: {ties}")
            break

        computer_choice = get_computer_choice()
        print(f"\nComputer chose: {computer_choice}")
        print(f"You chose: {user_choice}")

        result = decide_winner(computer_choice, user_choice)

        if result == "tie":
            print("It's a tie!")
            ties += 1
        elif result == "user":
            print(f"{user_choice.capitalize()} beats {computer_choice}! You win!")
            user_score += 1
        else:
            print(f"{computer_choice.capitalize()} beats {user_choice}! Computer wins!")
            computer_score += 1

        print(f"Score -> You: {user_score} | Computer: {computer_score} | Ties: {ties}")
        print("-" * 40)


# 5: Rock-Paper-Scissors (Two Player)

def get_player_choice(player_name):
    choice = input(f"{player_name}, enter a choice (rock, paper, scissors) or 'quit' to exit: ").lower().strip()
    while choice not in ['rock', 'paper', 'scissors', 'quit']:
        print("Invalid choice. Try again.")
        choice = input(f"{player_name}, enter a choice (rock, paper, scissors) or 'quit' to exit: ").lower().strip()
    return choice

def decide_winner(player1, choice1, player2, choice2):
    if choice1 == choice2:
        return "tie", f"Both chose {choice1}. It's a tie!"

    win_msgs = {
        ('rock', 'scissors'): "Rock crushes scissors!",
        ('scissors', 'paper'): "Scissors cut paper!",
        ('paper', 'rock'): "Paper covers rock!"
    }

    if (choice1, choice2) in win_msgs:
        return "player1", f"{win_msgs[(choice1, choice2)]} {player1} wins!"
    else:
        return "player2", f"{win_msgs[(choice2, choice1)]} {player2} wins!"

def play_rock_paper_scissors_user():
    print("Welcome to Two-Player Rock-Paper-Scissors with Score Tracking!")
    player1 = input("Player 1! Enter your name: ").strip()
    player2 = input("player 2! Enter your name: ").strip()

    score = {
        player1: 0,
        player2: 0,
        "ties": 0
    }

    while True:
        print("\n--- New Round ---")
        choice1 = get_player_choice(player1)
        if choice1 == 'quit':
            print(f"\n{player1} quit the game.")
            break

        choice2 = get_player_choice(player2)
        if choice2 == 'quit':
            print(f"\n{player2} quit the game.")
            break

        winner, message = decide_winner(player1, choice1, player2, choice2)
        print(message)

        if winner == "tie":
            score["ties"] += 1
        elif winner == "player1":
            score[player1] += 1
        else:
            score[player2] += 1

        print(f"\nScoreboard:\n{player1}: {score[player1]} | {player2}: {score[player2]} | Ties: {score['ties']}")
        print("-" * 40)

    print("\nFinal Scores:")
    print(f"{player1}: {score[player1]} | {player2}: {score[player2]} | Ties: {score['ties']}")
    print("Thanks for playing!")


# 6: Word Guessing Game

import random

def play_word_guessing():
    word_list = [
        "quiz", "data", "wizard", "incorrect", "secret",
        "box", "pen", "candle", "horse"
    ]

    total_correct = 0
    total_incorrect = 0
    total_games_played = 0

    print("Welcome to the Word Guessing Game!")
    print("Guess the secret word. You have 5 attempts each round.")
    print("Type 'quit' anytime to exit.\n")

    while True:
        secret_word = random.choice(word_list)
        max_attempt = 3
        attempt = 1
        correct = False

        print(f"\nNew word! It has {len(secret_word)} letters.")

        while attempt <= max_attempt:
            guess = input(f"\nEnter your guess ({attempt}/{max_attempt}) or type 'quit' to exit: ").strip().lower()

            if guess == 'quit':
                print("\nYou chose to quit the game.")
                print("\nFinal Scoreboard:")
                print(f"Total Games Played  : {total_games_played}")
                print(f"Correct Guesses     : {total_correct}")
                print(f"Incorrect Guesses   : {total_incorrect}")
                print("\nThanks for playing!")
                return

            if guess == secret_word:
                print("Correct! You've guessed the word.")
                correct = True
                total_correct += 1
                total_games_played += 1
                break
            else:
                print("Incorrect guess.")
                if attempt == 1:
                    print(f"Hint: The word starts with '{secret_word[0]}'")

            attempt += 1

        if not correct:
            print(f"\nYou've used all attempts. The word was: '{secret_word}'.")
            total_incorrect += 1
            total_games_played += 1


class GameMenu:
    @staticmethod
    def list():
        print("\n=== Game Menu ===")
        print("1. Play Riddle Puzzle")
        print("2. Play Arithmetic Puzzle")
        print("3. Play General Knowledge Quiz")
        print("4. Play Rock-Paper-Scissor (Computer)")
        print("5. Play Rock-Paper-Scissor (User)")
        print("6. Play Word-Guessing Game")
        print("7. Exit")
        print("Enter a number between 1 and 7.")


def main():
    while True:
        try:
            GameMenu.list()
            choice = input("Enter your choice: ")
            if choice == "1":
                riddle_game()
            elif choice == "2":
                arithmetic_puzzle()
            elif choice == "3":
                play_gk_quiz()
            elif choice == "4":
                play_rock_paper_scissors_computer()
            elif choice == "5":
                play_rock_paper_scissors_user()
            elif choice == "6":
                play_word_guessing()
            elif choice == "7":
                print("Exiting the game. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 7.")
        except Exception as e:
            print(f"An error occurred: {e}")

        again = input("Do you want to play another game? (Y/N): ")
        if again.lower() != "y":
            print("Thank you for playing!")
            break

main()

```
